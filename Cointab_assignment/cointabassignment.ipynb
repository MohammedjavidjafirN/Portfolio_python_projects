```python
def X_calulatedcharges(df3):
    c=0
    
    if(df3['Zone_X']=="b"):
        if(df3['Type of Shipment']=="Forward charges"):
            if(df3['charges']=='Fix'):
                c = 1*33
            else:
                c = 1*33 + (df3["No of slabs_x"]-1)*28.3
        
        else:
            if(df3['charges']=='Fix'):
                c= 1*20.5
            else:
                c=1*20.5 + (df3["No of slabs_x"]-1)*28.3        
            
    elif(df3['Zone_X']=="d"):
        if(df3['Type of Shipment']=="Forward charges"):
            if(df3['charges']=='Fix'):
                c = 1*45.4
            else:
                c=1*45.4 + (df3["No of slabs_x"]-1)*44.8
        else:
            if(df3['charges']=='Fix'):
                c = 1*41.3
            else:
                c = 1*41.3 + (df3["No of slabs_x"]-1)*44.8               
    
    elif(df3['Zone_X']=="e"):
        if(df3['Type of Shipment']=="Forward charges"):
            if(df3['charges']=='Fix'):
                c = 1*56.6
            else:
                c = 1*56.6 + (df3["No of slabs_x"]-1)*55.5

        else:
            if(df3['charges']=='Fix'):
                c = 1*50.7
            else:
                c = 1*50.7 + (df3["No of slabs_x"]-1)*55.5
    
    return c  
```

```python
df3['Expected charge_X'] = df3.apply(X_calulatedcharges , axis=1)
df3
```

```python
df3["Differnce in amount"]=df3['Expected charge_X']-df3['Billing Amount (Rs.)']
df3
```

```python
##Orders_overcharged=(df3['Differnce in amount']==0).sum()
#Orders_overcharged

over=(df3["Differnce in amount"]< 0).sum()
exact=(df3["Differnce in amount"]==0).sum()
under=(df3["Differnce in amount"] > 0).sum()
print(over,exact,under)
```
